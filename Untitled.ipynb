```python
import os
import lmdb 
import cv2
import numpy as np
import pandas as pd
```

```python
df = pd.read_pickle('../datamartV1.pkl')
```

```python
def writeCache(env, cache):
    with env.begin(write=True) as txn:
        for k, v in cache.iteritems():
            txn.put(k, v)
```

```python
def read_text(path):
    with open(path) as f:
        text = f.read()
    text = text.strip()
    return text
```

```python
df['Amount_hw'] = df.Amount_hw.apply(lambda x : x.encode())
```

```python
text_file = df.Amount_hw.values.tolist()
image_file = df.root.values.tolist()
outputPath ='/image/Chen/lmdb/'
```

```python
env = lmdb.open(outputPath, map_size=1099511627776)
```

```python
# env = lmdb.open(
#             outputPath,
#             max_readers=1,
#             readonly=False,
#             lock=False,
#             readahead=False,
#             meminit=False)
```

```python
cnt = 1
cache = {}
nSamples = len(image_file)
for i in range(nSamples):
    imagePath = image_file[i]
    label = text_file[i]
    if not os.path.exists(imagePath):
        print('%s does not exist' % imagePath)
        continue
    with open(imagePath,'rb') as f:
        image = f.read()
    imageKey = 'image-%09d' % cnt
    labelKey = 'label-%09d' % cnt
    cache[imageKey] = image
    cache[labelKey] = label
    ##########

    if cnt % 1 == 0:
        with env.begin(write=True) as txn:
            for k, v in zip(cache.keys(),cache.values()):
                txn.put(k.encode(), v)
        cache = {}
        print('Written %d / %d' % (cnt, nSamples))
    cnt += 1
nSamples = cnt-1
```

```python
cache['num-samples'] = str(nSamples)

with env.begin(write=True) as txn:
    for k, v in zip(cache.keys(),cache.values()):
        txn.put(k.encode(), v.encode())
print('Created dataset with %d samples' % nSamples)
```
